```python
from transformers import pipeline
# ---------- Text Summarization ----------
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
article = """Generative AI refers to a class of artificial intelligence models capable of
producing new content such as text, images, audio, and video. Large Language Models (LLMs)
such as GPT and LLaMA are trained on massive text corpora and can perform a wide range of
natural language tasks including translation, summarization, and question answering. These
models are increasingly being deployed in industry applications ranging from customer support
to software development, transforming how humans interact with machines."""
summary = summarizer(article, max_length=45, min_length=20, do_sample=False)
print("Summary:\n", summary[0]["summary_text"])
# ---------- Question Answering ----------
qa = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = article
question = "What are Large Language Models trained on?"
answer = qa(question=question, context=context)
print("\nQuestion:", question)
print("Answer:", answer["answer"], "| Confidence:", round(answer["score"], 3))
```

```python
from transformers import pipeline

# Load summarization model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Get input from user
print("Enter the text to summarize:")
text = input()

# Generate summary
summary = summarizer(
    text,
    max_length=50,
    min_length=20,
    do_sample=False
)

# Print summary
print("\nSummary:")
print(summary[0]["summary_text"])
```

```python
from transformers import pipeline

# ---------------- Summarization ----------------
summarizer = pipeline(
    "summarization",
    model="google/pegasus-xsum",
    framework="pt"
)

# ---------------- Question Answering ----------------
qa = pipeline(
    "question-answering",
    model="deepset/roberta-base-squad2",
    framework="pt"
)

# ---------------- User Input ----------------
article = input("Enter the text:\n")
question = input("Enter your question:\n")

# ---------------- Summarization ----------------
summary = summarizer(
    article,
    max_length=50,
    min_length=20,
    do_sample=False
)

print("\nSummary:")
print(summary[0]["summary_text"])

# ---------------- Question Answering ----------------
answer = qa(question=question, context=article)

print("\nQuestion:", question)
print("Answer:", answer["answer"])
print("Confidence:", round(answer["score"], 3))
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")

text = "Generative AI is transforming the world."

tokens = tokenizer.tokenize(text)

print("Tokenizer Class:", tokenizer.__class__.__name__)
print("Tokens:", tokens)
```

```python
from transformers import AutoTokenizer, BertTokenizer

text = "Generative AI is transforming the world."

# BART Tokenizer (Byte-Level BPE)
bart_tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")

# BERT Tokenizer (WordPiece)
bert_tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

print("BART Tokens:")
print(bart_tokenizer.tokenize(text))

print("\nBERT Tokens:")
print(bert_tokenizer.tokenize(text))
```
